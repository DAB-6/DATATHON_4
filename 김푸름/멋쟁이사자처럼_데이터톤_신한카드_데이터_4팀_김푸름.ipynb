#파생변수 생성

```python
# 동업종_매장수(같은 상권내 업종별 매장 수 = 업종별 경쟁강도확인)
# 상권 클러스터 묶음 안에 업종별로 묶음
df['동업종_매장수'] = (
    df.groupby(['상권클러스터', '업종대분류'])['업종대분류']
      .transform('count'))
```

```python
#상권내_매장수( 같은 상권내 모든 업종 매장수 = 삼권 규모 확인)
df['상권내_매장수'] = (
    df.groupby('상권클러스터')['업종대분류']
      .transform('count'))
```

```python
df[['상권클러스터','업종대분류','동업종_매장수', '상권내_매장수']].head(15)
```

#업종 비중

```python
# 1) 클러스터×업종대분류별 매장 수
cluster_industry = (
    df.groupby(['상권클러스터', '업종대분류'])
      .size()
      .reset_index(name='매장수'))


cluster_industry['업종비중'] = (
    cluster_industry
    .groupby('상권클러스터')['매장수']
    .transform(lambda x: x / x.sum()))

cluster_industry.head(10)
```

```python
# 상권별 상위 3개 업종만 추출
top_industries = (
    cluster_industry
    .sort_values(['상권클러스터', '업종비중'], ascending=[True, False])
    .groupby('상권클러스터')
    .head(3)
)

top_industries.head(50)
```

```python
#각  상권의 평균 성과지표
cluster_metrics = (
    df.groupby('상권클러스터')[['객단가_구간','월매출_구간','월고객수_구간']]
      .mean()
      .reset_index()
)
cluster_metrics.head(15)
```

```python
industry_pivot = cluster_industry.pivot(index='상권클러스터', columns='업종대분류', values='업종비중').fillna(0)

# 객단가·매출·고객수 결합
cluster_profile = pd.merge(cluster_metrics, industry_pivot, on='상권클러스터', how='left')
cluster_profile.head(15)
```

#통계검증

```python
import statsmodels.api as sm
import pandas as pd

def ols_summary_by_industry(df, y_cols, x_col, group_col='업종대분류', min_n=200):
    results = []

    for cat, d in df.groupby(group_col):
        if d.shape[0] < min_n:
            continue  # 표본이 너무 적으면 제외

        for y_col in y_cols:
            X = sm.add_constant(d[[x_col]].astype(float))
            y = d[y_col].astype(float)

            model = sm.OLS(y, X, missing='drop').fit(cov_type='HC3')

            results.append({
                '업종': cat,
                '지표': y_col,
                '표본수': int(model.nobs),
                'R2(설명력)': round(model.rsquared, 4),

                'p값': round(model.pvalues[x_col], 4)
            })

    return pd.DataFrame(results).sort_values(['지표','R2(설명력)'], ascending=[True, False])

# 실행 예시
y_cols = ['객단가_구간', '월매출_구간', '월고객수_구간']
summary_df = ols_summary_by_industry(df, y_cols, x_col='상권포화도', group_col='업종대분류')

# 결과 미리보기
summary_df.head(15)
```

```python
import statsmodels.api as sm
import pandas as pd

def top3_by_cluster(df, y_col='월매출_구간', x_col='상권포화도',
                    cluster_col='상권클러스터', group_col='업종대분류', min_n=100):

    results = []

    for cluster, df_cluster in df.groupby(cluster_col):
        # 각 클러스터 내 업종별 표본 수 상위 3개 선정
        top3_cats = (
            df_cluster[group_col]
            .value_counts()
            .head(3)
            .index.tolist()
        )

        for cat in top3_cats:
            d = df_cluster[df_cluster[group_col] == cat]
            if len(d) < min_n:
                continue

            X = sm.add_constant(d[[x_col]].astype(float))
            y = d[y_col].astype(float)
            model = sm.OLS(y, X, missing='drop').fit(cov_type='HC3')

            results.append({
                '클러스터': cluster,
                '업종': cat,
                '표본수': int(model.nobs),
                'R2(설명력)': round(model.rsquared, 4),
                'p값': round(model.pvalues[x_col], 4)
            })

    df_res = pd.DataFrame(results)
    return df_res.sort_values(['클러스터', 'R2(설명력)'], ascending=[True, False]).reset_index(drop=True)

# 실행
top3_df = top3_by_cluster(df, y_col='월매출_구간')
top3_df.head(30)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product

# --------------------------------
# 1️⃣ 데이터 로드 및 전처리
# --------------------------------

df["개설일"] = pd.to_datetime(df["개설일"], errors="coerce")
df["폐업일"] = pd.to_datetime(df["폐업일"], errors="coerce")
df["기준년월"] = pd.to_datetime(df["기준년월"], errors="coerce")

cluster_col = "상권클러스터"
id_col = "가맹점구분번호"
start, end = pd.Timestamp("2023-01-01"), pd.Timestamp("2024-12-31")

# --------------------------------
# 2️⃣ 상권 × 월별 운영 중 매장 수 계산
# --------------------------------
months = pd.period_range(start=start, end=end, freq="M").to_timestamp()
clusters = sorted(df[cluster_col].dropna().unique().tolist())
panel = pd.DataFrame(list(product(clusters, months)), columns=[cluster_col, "월"])

def count_active_stores(cluster, month):
    temp = df[df[cluster_col] == cluster]
    return temp[
        (temp["개설일"] <= month) &
        ((temp["폐업일"].isna()) | (temp["폐업일"] >= month))
    ][id_col].nunique()

panel["운영매장수"] = panel.apply(lambda x: count_active_stores(x[cluster_col], x["월"]), axis=1)
panel["변동률(%)"] = panel.groupby(cluster_col)["운영매장수"].pct_change() * 100

# --------------------------------
# 3️⃣ 성장률 요약 (상권별 평균 변동률)
# --------------------------------
growth_df = (
    panel.groupby(cluster_col)["변동률(%)"]
    .mean()
    .reset_index()
    .rename(columns={"변동률(%)": "평균변동률(%)"})
)

# --------------------------------
# 4️⃣ 내부 변동성 계산 (상대변동성 = 개폐업수 ÷ 전체매장수)
# --------------------------------
open_df = df.dropna(subset=["개설일"]).query("@start <= 개설일 <= @end")
close_df = df.dropna(subset=["폐업일"]).query("@start <= 폐업일 <= @end")

open_counts = (open_df.drop_duplicates([cluster_col, id_col])
                        .groupby(cluster_col)[id_col].nunique()
                        .reset_index(name="개업수"))
close_counts = (close_df.drop_duplicates([cluster_col, id_col])
                         .groupby(cluster_col)[id_col].nunique()
                         .reset_index(name="폐업수"))

activity = pd.merge(open_counts, close_counts, on=cluster_col, how="outer").fillna(0)
activity["총변동"] = activity["개업수"] + activity["폐업수"]

store_count = df.groupby(cluster_col)[id_col].nunique().reset_index(name="전체매장수")
stability_df = pd.merge(activity, store_count, on=cluster_col, how="left")
stability_df["상대변동성(%)"] = (stability_df["총변동"] / stability_df["전체매장수"]) * 100

# --------------------------------
# 5️⃣ 성장률 + 안정성 결합
# --------------------------------
merged = pd.merge(growth_df, stability_df, on=cluster_col, how="inner")

# --------------------------------
# 6️⃣ 시각화: 성장률 vs 안정성 (스캐터)
# --------------------------------
plt.figure(figsize=(10,7))
sns.scatterplot(
    data=merged,
    x="평균변동률(%)",
    y="상대변동성(%)",
    hue="상대변동성(%)",
    palette="coolwarm",
    size="상대변동성(%)",
    sizes=(40,300),
    alpha=0.8,
    edgecolor="white"
)

for i, row in merged.iterrows():
    plt.text(
        row["평균변동률(%)"],           # X좌표
        row["상대변동성(%)"],           # Y좌표
        str(row[cluster_col]),          # 표시할 텍스트 (상권 번호)
        fontsize=9, color="black", ha="center", va="center",
        weight="bold"
    )

plt.title("빠르게 성장하는 상권일수록 불안정할까? (성장률 vs 변동성)", fontsize=14)
plt.xlabel("상권 성장률 (전월 대비 평균 변동률 %)")
plt.ylabel("상권 변동성 (상권 내 교체율 %)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product
from sklearn.preprocessing import MinMaxScaler


start, end = pd.Timestamp("2023-01-01"), pd.Timestamp("2024-12-31")

df["개설일"] = pd.to_datetime(df["개설일"], errors="coerce")
df["폐업일"] = pd.to_datetime(df["폐업일"], errors="coerce")
df["기준년월"] = pd.to_datetime(df["기준년월"], errors="coerce")

cluster_col = "상권클러스터"
id_col = "가맹점구분번호"

# 2️⃣ 상권×월별 운영 매장수 & 변동률(성장률)
months = pd.period_range(start=start, end=end, freq="M").to_timestamp()
clusters = sorted(df[cluster_col].dropna().unique().tolist())
panel = pd.DataFrame(list(product(clusters, months)), columns=[cluster_col, "월"])

def count_active_stores(cluster, month):
    sub = df[df[cluster_col] == cluster]
    active = sub[(sub["개설일"] <= month) & ((sub["폐업일"].isna()) | (sub["폐업일"] >= month))]
    return active[id_col].nunique()

panel["운영매장수"] = panel.apply(lambda x: count_active_stores(x[cluster_col], x["월"]), axis=1)
panel["변동률(%)"] = panel.groupby(cluster_col)["운영매장수"].pct_change() * 100

growth_df = (
    panel.groupby(cluster_col, as_index=False)["변동률(%)"]
    .mean()
    .rename(columns={"변동률(%)": "평균변동률(%)"})
)

# 3️⃣ 상권 변동성(%) 계산 = (개업 + 폐업) / 전체매장 ×100
open_df = df.dropna(subset=["개설일"]).query("@start <= 개설일 <= @end").copy()
close_df = df.dropna(subset=["폐업일"]).query("@start <= 폐업일 <= @end").copy()

open_counts = (
    open_df.drop_duplicates([cluster_col, id_col])
    .groupby(cluster_col)[id_col].nunique()
    .reset_index(name="개업수")
)
close_counts = (
    close_df.drop_duplicates([cluster_col, id_col])
    .groupby(cluster_col)[id_col].nunique()
    .reset_index(name="폐업수")
)

activity = pd.merge(open_counts, close_counts, on=cluster_col, how="outer").fillna(0)
activity["총변동"] = activity["개업수"] + activity["폐업수"]

store_count = df.groupby(cluster_col)[id_col].nunique().reset_index(name="전체매장수")
stability_df = pd.merge(activity, store_count, on=cluster_col, how="left")

# 상권 변동성(%)
stability_df["상권변동성(%)"] = (stability_df["총변동"] / stability_df["전체매장수"]) * 100

# 4️⃣ 내부 안정성 온도(0~100) = 100 - minmax(변동성)
scaler = MinMaxScaler(feature_range=(0, 100))
stability_df["변동성_norm(0~100)"] = scaler.fit_transform(stability_df[["상권변동성(%)"]])
stability_df["내부안정성온도(0~100)"] = 100 - stability_df["변동성_norm(0~100)"]

# 5️⃣ 성장률 + 안정성 결합
merged = pd.merge(growth_df, stability_df, on=cluster_col, how="inner")

# 6️⃣ 스캐터 시각화 (버전 선택)
# version = "volatility"  # 내부 변동성 보기 (불안정이 위로)
version = "stability"   # 내부 안정성 보기 (안정이 위로)

plt.figure(figsize=(10, 7))

if version == "volatility":
    sns.scatterplot(
        data=merged,
        x="평균변동률(%)",
        y="상권변동성(%)",
        hue="상권변동성(%)",
        palette="coolwarm",
        size="상권변동성(%)",
        sizes=(40, 300),
        alpha=0.85,
        edgecolor="white"
    )
    plt.ylabel("상권 변동성(%) (교체율, 불안정↑)")
    plt.title("상권 성장률 vs 상권 변동성 (불안정성 기준)", fontsize=14)

else:
    sns.scatterplot(
        data=merged,
        x="평균변동률(%)",
        y="내부안정성온도(0~100)",
        hue="내부안정성온도(0~100)",
        palette="coolwarm_r",
        size="내부안정성온도(0~100)",
        sizes=(40, 300),
        alpha=0.85,
        edgecolor="white"
    )
    plt.ylabel("내부 안정성 온도(0~100) (안정↑)")
    plt.title("상권의 지속 가능성 진단", fontsize=14)

# 상권 번호 라벨 추가
for _, row in merged.iterrows():
    plt.text(
        row["평균변동률(%)"],
        row["내부안정성온도(0~100)"] if version == "stability" else row["상권변동성(%)"],
        str(int(row["상권클러스터"])),
        fontsize=9, color="black", ha="center", va="center", weight="bold"
    )

plt.axvline(x=merged["평균변동률(%)"].median(), color="gray", linestyle="--", alpha=0.6)
if version == "volatility":
    plt.axhline(y=merged["상권변동성(%)"].median(), color="gray", linestyle="--", alpha=0.6)
else:
    plt.axhline(y=merged["내부안정성온도(0~100)"].median(), color="gray", linestyle="--", alpha=0.6)

plt.xlabel("상권 성장률 (전월 대비 평균 변동률 %)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 결과 확인
print("\n📊 결과 예시")
print(merged[[cluster_col, "평균변동률(%)", "상권변동성(%)", "내부안정성온도(0~100)"]].head())
```

```python
start = pd.Timestamp("2023-01-01")
end   = pd.Timestamp("2024-12-31")

df = df.copy()
df["개설일"] = pd.to_datetime(df["개설일"], errors="coerce")
df["폐업일"] = pd.to_datetime(df["폐업일"], errors="coerce")
df["기준년월_dt"] = pd.to_datetime(df.get("기준년월_dt"), errors="coerce")  # 없으면 NaT

# ===== 2) 개/폐업 월별 집계 (size) =====
df_open  = df[(df["개설일"] >= start) & (df["개설일"] <= end)].copy()
df_close = df[(df["폐업일"] >= start) & (df["폐업일"] <= end)].copy()
df_open["event_month"]  = df_open["개설일"].dt.to_period("M").dt.to_timestamp()
df_close["event_month"] = df_close["폐업일"].dt.to_period("M").dt.to_timestamp()

open_size  = df_open.groupby(["상권클러스터","event_month"]).size().rename("개업수")
close_size = df_close.groupby(["상권클러스터","event_month"]).size().rename("폐업수")

trend = pd.concat([open_size, close_size], axis=1).fillna(0).reset_index()
trend["개업수"] = trend["개업수"].astype(int)
trend["폐업수"] = trend["폐업수"].astype(int)
trend["순변동"] = trend["개업수"] - trend["폐업수"]

# 월별 상권규모(분모): 해당 월 존재한 고유 점포 수(있으면 권장), 없으면 전체 기간 상권규모
if "기준년월_dt" in df.columns:
    cluster_size_monthly = (
        df[(df["기준년월_dt"] >= start) & (df["기준년월_dt"] <= end)]
        .groupby(["상권클러스터","기준년월_dt"])["가맹점구분번호"]
        .nunique()
        .reset_index()
        .rename(columns={"기준년월_dt":"event_month","가맹점구분번호":"상권규모"})
    )
    trend = trend.merge(cluster_size_monthly, on=["상권클러스터","event_month"], how="left")
else:
    cluster_size_all = (
        df.groupby("상권클러스터")["가맹점구분번호"].nunique()
        .rename("상권규모").reset_index()
    )
    trend = trend.merge(cluster_size_all, on="상권클러스터", how="left")

trend["상권규모"] = trend["상권규모"].fillna(0).astype(int)
trend["순변동비율"] = np.where(trend["상권규모"] > 0, trend["순변동"] / trend["상권규모"], np.nan)

# ===== 3) 상권별 표준편차(변동성) 계산: 절대값 vs 비율 =====
vol_abs = (
    trend.groupby("상권클러스터")["순변동"]
         .std(ddof=0)
         .reset_index()
         .rename(columns={"순변동":"변동성_std(절대값)"})
)

vol_ratio = (
    trend.groupby("상권클러스터")["순변동비율"]
         .std(ddof=0)
         .reset_index()
         .rename(columns={"순변동비율":"변동성_std(비율)"})
)

vol = vol_abs.merge(vol_ratio, on="상권클러스터", how="left")

# ===== 4) 공용 함수: 중앙값 기준 상/하위 그룹 요약 출력 =====
def print_group_stats(df, colname, title):
    median_val = df[colname].median()
    upper = df[df[colname] >= median_val].copy()
    lower = df[df[colname] <  median_val].copy()
    upper_mean = upper[colname].mean()
    lower_mean = lower[colname].mean()

    print(f"\n=== {title} ===")
    print(f"🔸 중앙값 : {median_val:.6f}")
    print(f"🔹 상위 평균 : {upper_mean:.6f}  (n={len(upper)})")
    print(f"🔹 하위 평균 : {lower_mean:.6f}  (n={len(lower)})")

    print("\n[상위 그룹 상권]")
    print(upper[["상권클러스터", colname]].sort_values(colname, ascending=False).to_string(index=False))
    print("\n[하위 그룹 상권]")
    print(lower[["상권클러스터", colname]].sort_values(colname, ascending=False).to_string(index=False))

# ===== 5) 두 기준 각각 출력 =====
print_group_stats(vol, "변동성_std(절대값)", "변동성(순변동 표준편차) — 절대값 기준")
print_group_stats(vol, "변동성_std(비율)",  "변동성(순변동 표준편차) — 비율 기준")
```

```python
def make_multi_samples(df, target_cols, window=12,
                       extra_cols=None, store_col='가맹점구분번호',
                       date_col='기준년월', dropna=True):
    X_list, y_list, idx = [], [], []
    extra_cols = extra_cols or []  # 추가 피처 리스트 (예: ['sin_month', 'cos_month'])

    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values([store_col, date_col]).reset_index(drop=True)

    for sid, g in df.groupby(store_col):
        vals = g[target_cols].values
        months = g[date_col].values
        extras = g[extra_cols].values if extra_cols else None

        for i in range(len(vals) - window):
            X_window = vals[i:i+window].flatten()  # 12개월 x 타깃 수
            y_next = vals[i+window]               # 다음 달 타깃
            extra_feat = extras[i+window] if extra_cols else []

            if np.isnan(X_window).any() or np.isnan(y_next).any():
                continue

            X_list.append(np.concatenate([X_window, extra_feat]))
            y_list.append(y_next)
            idx.append((sid, pd.to_datetime(months[i+window])))

    return np.array(X_list), np.array(y_list), idx

extra_cols = ['sin_month', 'cos_month']

target_cols = ['신규고객_비중_cleaned', '재방문고객_비중_cleaned', '유동인구고객_비중_cleaned']
X, y, idx = make_multi_samples(df, target_cols, window=12)
print("X shape:", X.shape)
print("y shape:", y.shape)  # (샘플수, 3)
print("idx sample:", idx[:3])
```

```python
import numpy as np
import pandas as pd

target_months = pd.PeriodIndex([dt for (_, dt) in idx], freq='M')
uniq_months = np.sort(target_months.unique())

# 최근 4개월 테스트로 지정
N_TEST_MONTHS = 4
test_months = uniq_months[-N_TEST_MONTHS:]
is_test = target_months.isin(test_months)

# 훈련
X_train, X_test = X[~is_test], X[is_test]
y_train, y_test = y[~is_test], y[is_test]

print("Train:", X_train.shape, " Test:", X_test.shape)
print("Test months:", list(map(str, test_months)))

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rf = MultiOutputRegressor(RandomForestRegressor(
    n_estimators=300, random_state=42, n_jobs=-1
))
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
```

```python
def smape(y_true, y_pred, eps=1e-8):
    den = (np.abs(y_true) + np.abs(y_pred)).clip(min=eps)
    return np.mean(np.abs(y_pred - y_true) / den) * 100

metrics = {}
for i, col in enumerate(target_cols):
    rmse = mean_squared_error(y_test[:, i], pred[:, i], squared=False)
    mae = mean_absolute_error(y_test[:, i], pred[:, i])
    r2 = r2_score(y_test[:, i], pred[:, i])
    s = smape(y_test[:, i], pred[:, i])
    metrics[col] = {'RMSE': rmse, 'MAE': mae, 'R2': r2, 'sMAPE(%)': s}

pd.DataFrame(metrics)
```
