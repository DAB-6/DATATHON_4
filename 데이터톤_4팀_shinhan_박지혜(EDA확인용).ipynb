### 데이터셋 1 전처리 및 파생변수 생성

```python
# 개설일, 폐업일을 datetime 형식으로 변환
df1['개설일'] = pd.to_datetime(df1['개설일'], format='%Y%m%d')
df1['폐업일'] = pd.to_datetime(df1['폐업일'], format='%Y%m%d', errors='coerce') # 오류 발생 시 NaT으로 처리
```

```python
# '폐업여부' 파생변수 생성 (폐업일이 NaT가 아니면 1, NaT이면 0)
df1['폐업여부'] = df1['폐업일'].notna().astype(int)
```

```python
# '운영기간' 파생변수 생성 (폐업 가맹점: 폐업일 - 개설일, 운영중 가맹점: 2024-12-31 - 개설일)
end_date = pd.to_datetime('2024-12-31')
df1['운영기간'] = (df1['폐업일'].fillna(end_date) - df1['개설일']).dt.days
```

```python
df1.head()
```

### 기술통계 및 시각화

```python
import numpy as np
import seaborn as sns
```

```python
# 폐업여부 분포 -> 폐업했으면 1, 안했으면 0
sns.countplot(x='폐업여부', data=df1)
plt.title('폐업여부 분포')
plt.show()
```

## 데이터셋 1,2 병합 및 분석

```python
# df1_to_merge = df1[['가맹점구분번호', '폐업여부', '운영기간']]
df = pd.merge(df2, df1, on='가맹점구분번호', how='left')
```

```python
df.head(5)
```

```python
df.shape
```

### 병합된 데이터 전처리

```python
#오기입 컬럼명 변경
df.rename(columns={
    '배달매출비율': '취소율_구간',
    '배달매출비율.1': '배달매출비율',
    '12개월_업종대비_평균고객수_비율':'12개월_업종대비_평균매출건수_비율',
    '12개월_상권내_가맹점_비중':'12개월_동일업종_폐업가맹점_비중'},
inplace=True)
```

```python
df.head(5)
```

```python
# 각 구간의 실제 구간 숫자만 보이게 변환 > % 상세부분 잘라내기
def convert_range_to_numeric(column):
    numeric_part = column.astype(str).str.split('_').str[0]
    numeric_part = numeric_part.fillna('0')
    return numeric_part.astype(int)

range_columns = ['운영개월수_구간', '월매출_구간', '월매출건수_구간', '월고객수_구간', '객단가_구간', '취소율_구간']

for col in range_columns:
    if col == '취소율_구간':
        df[col] = df[col].fillna('0')
    df[col] = convert_range_to_numeric(df[col])

df.replace(-999999.9, np.nan, inplace=True)
```

```python
df.head(5)
```

```python
categorical_cols = ['운영개월수_구간', '월매출_구간', '월매출건수_구간', '월고객수_구간', '객단가_구간', '취소율_구간']
for col in categorical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))

    # Non-closed restaurants
    sns.countplot(data=df[df['폐업여부'] == 0], x=col, ax=axes[0])
    axes[0].set_title(f'{col} (미폐업)')
    axes[0].tick_params(axis='x', rotation=45)

    # Closed restaurants
    sns.countplot(data=df[df['폐업여부'] == 1], x=col, ax=axes[1])
    axes[1].set_title(f'{col} (폐업)')
    axes[1].tick_params(axis='x', rotation=45)

    plt.suptitle(f'{col} 별 폐업여부 비교')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
continuous_cols = ['배달매출비율', '12개월_업종대비_평균매출_비율', '12개월_업종대비_평균매출건수_비율', '12개월_업종내_매출_백분위', '12개월_상권내_매출_백분위', '12개월_동일업종_폐업가맹점_비중', '12개월_동일상권_폐업가맹점_비중', '운영기간']
for col in continuous_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='폐업여부', y=col)
    plt.title(f'폐업여부에 따른 {col} 분포')
    plt.show()
```
